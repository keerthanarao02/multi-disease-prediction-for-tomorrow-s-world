```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
import pickle
```

```python
lung_cancer_dataset = pd.read_csv('/content/survey lung cancer.csv')
```

```python
lung_cancer_dataset.head()
```

```python
lung_cancer_dataset['GENDER'] = lung_cancer_dataset['GENDER'].map({'M': 0, 'F': 1})
```

```python
lung_cancer_dataset.head()
```

```python
X_lung = lung_cancer_dataset.drop(columns='LUNG_CANCER', axis=1)
Y_lung = lung_cancer_dataset['LUNG_CANCER']
```

```python
X_train_lung, X_test_lung, Y_train_lung, Y_test_lung = train_test_split(X_lung, Y_lung, test_size=0.2, stratify=Y_lung, random_state=2)
```

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train_lung, Y_train_lung)
```

```python
lung_train_prediction = classifier.predict(X_train_lung)
lung_train_accuracy = accuracy_score(lung_train_prediction, Y_train_lung)
print('Accuracy score on the training data:', lung_train_accuracy)
```

```python
input_data_lung = (0,69,1,2,2,1,1,2,1,2,2,2,2,2,2)  # Provide the feature values for prediction
input_data_as_numpy_array_lung = np.asarray(input_data_lung)
input_data_reshaped_lung = input_data_as_numpy_array_lung.reshape(1, -1)

# Make a prediction using the loaded model
prediction_lung = classifier.predict(input_data_reshaped_lung)

# Display the prediction result
if prediction_lung[0] == 0:
    print('The person does not have lung cancer.')
else:
    print('The person has lung cancer.')
```

```python
# Save the lung cancer model to a file
filename = 'lung_cancer_model.sav'
pickle.dump(classifier, open(filename, 'wb'))
```
